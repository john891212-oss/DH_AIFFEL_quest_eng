지금 설명 드리려는건 쉽게말해서 오류발생에 대비하는 안전장치라고 보면 됩니다.
보통 코드는 코딩을 하고 오류가 발생하면 그냥 그자리에서 '아 몰랑' 그러면서 그자리에서 주저 앉습니다. 이런 걸 방지하기 위해 사용되는 코드들이 한두개는 아니겠지만 오늘 우리가 배울건 try, except, else, finally  이 4개입니다.


각각 이해하기 쉽게 설명하자면


*   try : 일단 진행시켜 또는 일단 잡숴봐 처럼 결과가 예측되지 않지만 '일단' 가자는 느낌으로 코드를 여기에 작성합니다.

*   except : 이제 진행시켰거나 잡숴봤는데 망했을때, 작동하는 안전장치입니다.

*   else : 반대로 진행했는데 성공했다 또는 먹었는데 좋았다 그러니까 코드적으로는 오류 없이 성공햇을때 실행이 되는겁니다.

*   finally : 이건 이제 위에 죽이되든 밥이되든 무조건 나는 마무리를 할거다 라는 느낌으로 무조건 실행되는 코드입니다.

```python
# 문제 1번
# 해당 코드를 실행하면 어떤 결과가 나올까?

try:
    print('A')
    x = 1 / 0
    print('B')
except ZeroDivisionError:
    print('C')
finally:
    print('D')
```

해설  
자 위에서 설명한대로 try <- 시도 프린트 'A' 때문에  
A가 가장 먼저 올라옵니다.  
B는 어디갔냐? 'x = 1 / 0' 즉 1을 0으로 나누는 짓은 할 수 없는 행위입니다.     이걸 코드로 따로 작성해보면 'ZeroDivisionError:'   이런 에러가 발생합니다.  
그래서 코딩 특성상 아까 뭐랬죠? 오류가 발생하면 아 몰랑하고 주저 앉습니다.  
그래서 print('B')까지 도달하지를 못하니까 B는 출력이 안돼는 겁니다.  
그럼 C는? 익셉트 트라이에서 망해서 'ZeroDivisionError' 오류가 발생했으니까   백업으로 써둔 print('C')를 실행하는 겁니다.  
고로 A다음에 나오는 문자는 C가 되는겁니다.  
그리고 finally는 위에 설명해드렸다 싶이 코드가 죽이되는 밥이되든 나는 무조건 출력이 되겠다. 라는 느낌의 이라서 마지막의 print('D')가 출력되어서 답은

A  
C  
D  
이렇게 나오는겁니다.

```python
# 문제 2번
# 해당 코드를 실행하면 어떤 결과가 나올까?

try:
    print('A')
except:
    print('B')
else:
    print('C')
finally:
    print('D')
```

해설
역시 A가 먼저 올라옵니다. try로 프린트 'A'를 잡숴봤기 때문에 이상없이 'A'가 출력됩니다.  
그럼 오류안생겼는데 B는 어디갔나? 익셉트 뒤에 아무런 조건을 쓰지않았습니다.   츄라이에서 어떤 오류도 실패도 발생을 안했기때문에 익셉트 조건이 발동하지않아서 B는 재껴지는 겁니다.  
그리고 엘스, 트라이(A)는 성공했고 예외(B)는 건너뛰었으니까 백업인 엘스가 실행이 되는겁니다.  
그리고 마찬가지로 finally는 그냥 무조건 마지막을 장식하는 겁니다.  
고로  
A  
C  
D  
가 정답입니다.

```python
# 문제 3번
# 해당 코드를 실행하면 어떤 결과가 나올까?

def func():
    try:
        return '츄라이 츄라이 잡숴봐'
    finally:
        return '마지막은 무조건 나지'

print(func())
```

결국은 위의 두 문제에 대한 복습인데요.  
정답은 finally의 '마지막은 무조건 나지'입니다.  

코드를 해설해보자면, try에서 리턴 츄라이를 때렸지만 finally는 그냥 무조건 마지막을 장식하는 겁니다.  
위에 뭐가나왔든 마지막은 finally다. 라고 이해하시면됩니다.
